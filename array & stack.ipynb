```python
#INSERTION
arr=[10,20,30,40,50]
arr.insert(2,30)
print(arr)
```

```python
#traversel
arr=[10,20,30,40,50]
for element in arr:
    print(element)
```

```python
#DELETION
arr=[10,20,30,40,50]
arr.remove(30)
print(arr)
```

```python
#SEARCH
arr=[10,20,30,40,50]
key=30
if key in arr:
    print("Element Found")
else:
    print("Element not Found")
```

```python
#update
arr=[10,20,30,40,50]
arr[2]=35
print(arr)
```

```python
#sorting
arr=[40,10,50,20,20]
arr.sort()
print(arr)
```

```python
#accessing
arr=[40,10,50,20,20]
print(arr[0])
```
