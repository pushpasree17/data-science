```python
class myStack:
    # Constructor
    def __init__(self, cap):
        self.capacity = cap
        self.arr = [0] * self.capacity
        self.top = -1

    # Push operation
    def push(self, x):
        if self.top == self.capacity - 1:
            print("Stack Overflow")
            return
        self.top += 1
        self.arr[self.top] = x

    # Pop operation
    def pop(self):
        if self.top == -1:
            print("Stack Underflow")
            return -1
        val = self.arr[self.top]
        self.top -= 1
        return val

    # Peek operation
    def peek(self):
        if self.top == -1:
            print("Stack is Empty")
            return -1
        return self.arr[self.top]

    # Check if stack is empty
    def isEmpty(self):
        return self.top == -1

    # Check if stack is full
    def isFull(self):
        return self.top == self.capacity - 1


if __name__ == "__main__":

    # Get stack size
    size = int(input("Enter the size of the stack: "))

    # Create stack
    st = myStack(size)

    # Get number of elements
    n = int(input("Enter the number of elements to push: "))

    if n > size:
        print("Number of elements cannot exceed stack size.")
    else:
        # Push elements
        for i in range(n):
            x = int(input(f"Enter element {i+1}: "))
            st.push(x)

        # Pop one element
        print("Popped:", st.pop())

        # Display top element
        print("Top element:", st.peek())

        # Check if stack is empty
        print("Is stack empty:", "Yes" if st.isEmpty() else "No")

        # Check if stack is full
        print("Is stack full:", "Yes" if st.isFull() else "No")
```
